#### 1) Quais as principais características de linguagens orientadas para objetos?
Herança, polimorfismo e modularidade. 
#### 2) Quais os objetivos a serem alcançados com herança?
Diminui a reescrita de código repassando características gerais para 'filhos' de uma classe. 
#### 3) Qual a diferença entre herança de classes concretas e abstratas?
Uma classe abstrata específica métodos, mas cabe aos seus filhos implemetarem. Enquanto 
classes concretas implemetam métodos a serem herdados por seus filhos. 
#### 4) Entre herança e composição, porque atualmente se incentiva a segunda como forma de obter reuso de código?
Pois facilita a manuntenção de código e impede que métodos sejam e váriaveis sejam herdados de forma desconhecida
por classes.
#### 5) O que é programação orientada a aspectos (POA)?
No paradigma de programação orientada a objetos os interesses são divididos em diversas classes 
resposáveis por realizar suas partes. A modulação dessas classes é baseada nos interesses do projeto.
#### 6) Qual a diferença entre linguagens reflexivas e introspectivas?
Linguagens introspectivas são aquelas capazes de visualizar características de suas clases e métodos, enquanto
reflexivas são linguagens capazes alterar sua própria estrutura.
