#### 1) Quais as principais características de linguagens orientadas para objetos?
Herança, polimorfismo e modularidade. 
#### 2) Quais os objetivos a serem alcançados com herança?
Diminui a reescrita de código repassando características gerais para 'filhos' de uma classe. 
#### 3) Qual a diferença entre herança de classes concretas e abstratas?
Uma classe abstrata específica métodos, mas cabe aos seus filhos implemetarem. Enquanto 
classes concretas implemetam métodos a serem herdados por seus filhos. 
#### 4) Entre herança e composição, porque atualmente se incentiva a segunda como forma de obter reuso de código?
Pois facilita a manuntenção de código e impede que métodos sejam e váriaveis sejam herdados de forma desconhecida
por classes.
#### 5) O que é programação orientada a aspectos (POA)?
No paradigma de programação orientada a objetos os interesses são divididos em diversas classes 
resposáveis por realizar suas partes. A modulação dessas classes é baseada nos interesses do projeto.
#### 6) Qual a diferença entre linguagens reflexivas e introspectivas?
Linguagens introspectivas são aquelas capazes de visualizar características de suas clases e métodos, enquanto
reflexivas são linguagens capazes alterar sua própria estrutura. 

#### 7) Implemente um servidor socket em Python OO para manipulação de figuras geométricas em um grid gráfico em ambiente X. Seu servidor deve iniciar um grid de 5x5 células, como visto na figura abaixo, e esperar por requisições para criação, remoção e movimentação de figuras. Mais especificamente, o servidor deve reconhecer os seguintes comandos enviados por clientes socket:
* \+ id F C X Y: criar figura com identificador id (inteiro), forma F (s para quadrado e c para circulo), cor C (pelo menos, black, red, green e blue) e coordenadas X e Y (inteiros de 0 a 4).
* \- id: apagar figura identificada por id.
* m id X Y: mover figura identificada por id para coordenadas X e Y.

#### Os comandos podem opcionalmente não fazer referência explícita aos identificadores. Neste caso, identificadores adequados devem ser criados implicitamente.

* \+ F C X Y: criar figura com forma F, cor C e coordenadas X e Y.
* \- F C: apagar figura identificada por F C.
* m F C X Y: mover figura identificada por F C para coordenadas X e Y.

            self.w.create_line([(0, i), (w, i)])

def draw_circle(self, lin, col):
    x = col * self.cell_h
    y = lin * self.cell_w
    return self.w.create_oval(x + 10, y + 10,
        x + self.cell_w - 10, y + self.cell_h - 10,
        fill = 'blue', outline = '')

    def __init__(self, grid):
        Thread.__init__(self)
        self.grid = grid
        self.server = socket()
        self.server.bind((HOST, PORT))
        self.server.listen(5)
        self.client, addr = self.server.accept()
    def process_cmd(self, cmd):
        # + id shape(c|s) color lin col
        # - id
        # m id lin col
        # + shape(c|s) color lin col
        # - shape(c|s) color
        # m shape(c|s) color lin col
        reply = 'Done.\n'
        self.grid.draw_circle(2, 2)
        return reply
    def run(self):
        while True:
            try:
                text = self.client.recv(1024)
                if not text:
                    break
                reply = self.process_cmd(text)
                self.client.sendall(reply)

                break
        self.client.close()
if __name__ == '__main__':
